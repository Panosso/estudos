```python
# /content/sample_data
import pandas as pd
```

```python
# Loading CSV
game_status_df = pd.read_csv('/content/game_status.csv')
```

```python
#Todos os games com timeout
game_status_timeout = game_status_df[game_status_df['Como finalizou'] == 'timeout']

print(f"Quantidade de jogos com timeout: {game_status_timeout.shape[0]}")
```

```python
#Media turno de todas as partidas
game_status_turn = game_status_df['Numero de turnos'].mean(axis = 0)

print(f"Média de turnos por partida, contando timeout: {round(game_status_turn)}")
```

```python
#Média de turnos sem a finalização por timeout
game_status_turn_no_timeout = game_status_df[game_status_df['Como finalizou'] != 'timeout']['Numero de turnos'].mean(axis = 0)

print(f"Média de turnos por partida, sem timeout: {round(game_status_turn_no_timeout)}")
```

```python
winner_strategy = ['Impulsive', 'Random', 'Demanding', 'Cautious']

total = game_status_df['Estrategia Vencedor'].count()

highest_percentual = 0
highest_strategy_wr = ''

for strategy in winner_strategy:
  strategy_count = game_status_df['Estrategia Vencedor'][game_status_df['Estrategia Vencedor'] == strategy].count()
  percentual_strategy = round((strategy_count / total) * 100, 2)
  if percentual_strategy > highest_percentual:
    highest_percentual = percentual_strategy
    highest_strategy_wr = strategy
  print(f"A estratégia '{strategy}'' tem uma porcentagem de vitória de: {round(percentual_strategy, 2)}% \n")

print(f"A maior taxa de vitória é: {highest_percentual}% estratégia '{highest_strategy_wr}'")
```
