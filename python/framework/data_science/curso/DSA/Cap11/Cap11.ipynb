```python
alpha = 0.7
phi_ext = 2 * np.pi * 0.5

def color_map(phi_m, phi_p):
    return (alpha -2 * np.cos(phi_p)*np.cos(phi_m) - alpha * np.cos(phi_ext - 2*phi_p))

# Mais dados
phi_m = np.linspace(0, 2*np.pi, 100)
phi_p = np.linspace(0, 2*np.pi, 100)
X,Y = np.meshgrid(phi_p, phi_m)
Z = color_map(X, Y).T
```

```python
# Cria a figura
fig = plt.figure(figsize = (14,6))

# Adiciona o subplot 1 com proje√ß√£o 3d
ax = fig.add_subplot(1, 2, 1, projection = '3d')
p = ax.plot_surface(X, Y, Z, rstride=4, cstride=4, linewidth=0)

# Adiciona o subplot 2 com proje√ß√£o 3d
ax = fig.add_subplot(1, 2, 2, projection = '3d')
p = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='Accent', linewidth=0, antialiased=False)

# Cria a barra de cores como legenda
cb = fig.colorbar(p, shrink=0.5)
```

```python
# Valores rand√¥micos
np.random.seed(42)
n = 1000
pct_smokers = 0.2

# Vari√°veis
flag_fumante = np.random.rand(n) < pct_smokers
idade = np.random.normal(40, 10, n)
altura = np.random.normal(170, 10, n)
peso = np.random.normal(70, 10, n)

# Dataframe
dados = pd.DataFrame({'altura': altura, 'peso': peso, 'flag_fumante': flag_fumante})

# Cria os dados para a vari√°vel flag_fumante
dados['flag_fumante'] = dados['flag_fumante'].map({True: 'Fumante', False: 'N√£o Fumante'})
```

```python
# Style
sea.set(style = "ticks")

# lmplot
sea.lmplot(x = 'altura', 
           y = 'peso', 
           data = dados, 
           hue = 'flag_fumante', 
           palette = ['tab:blue', 'tab:orange'], 
           height = 7)

# Labels e t√≠tulo
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.title('Rela√ß√£o Entre Altura e Peso de Fumantes e N√£o Fumantes')

# Remove as bordas
sea.despine()

# Show
plt.show()
```
