```python
# Exercício 1 - Crie uma estrutura que pergunte ao usuário qual o dia da semana. Se o dia for igual a Domingo ou 
# igual a sábado, imprima na tela "Hoje é dia de descanso", caso contrário imprima na tela "Você precisa trabalhar!"

dia = (input('digite um dia da semana: '))

if dia == 'sabado' or dia == 'domingo':
    print('Hoje é dia de descanso')
else:
    print('Você precisa trabalhar')



#Solução Machado
print(10 * '------')
print('Solução Machado')

#Validamos se o valor da variavel dia está dentro desses valores informados, caso esteja, o fluxo de código entrará no if.
#Adicionei 'Sexta' só pra vc ver como podem ser qualquer valor, desde que esteja na tupla ele entra no if

if dia in ('sabado', 'domingo', 'sexta'):
    print('Hoje é dia de descanso')
else:
    print('Você precisa trabalhar')

#Eis aqui um desafio, se o usuário digitar 'sAbAdO' ou 'DOMIngo' a mensagem impressa será 'Você precisa trabalhar'. 
#Para resolver isso pesquise sobre o método lower()
```

```python
# Exercício 2 - Crie uma lista de 5 frutas e verifique se a fruta 'Morango' faz parte da lista

lista1 = ['chocolate', 'maça', 'mamao', 'caqui', 'goiaba']

for frutas in lista1:
  if frutas == 'morango':
    print('morango faz parte da lista!')
  else:
    print('morango não faz parte da lista!')


#Solução Machado
print(10 * '------')
print('Solução Machado')
#Vamos melhorar esses nomes de variáveis, lembrando que quando estamos programando, nosso código será lido por mais de um programador
#Se eu criar uma variável com um nome "gaisfgaushduias", ela não consegui passar nenhuma informação de quais valores ela contem
frutas = ['chocolate', 'maça', 'mamao', 'caqui', 'goiaba']

if 'morango' in frutas:
  print('morango faz parte da lista!')
else:
  print('morango não faz parte da lista!')

```

```python
# Exercício 3 - Crie uma tupla de 4 elementos, multiplique cada elemento da tupla por 2 e guarde os resultados em uma 
# lista
tupla= (10, 10, 10, 10)
lista= []
for i in tupla:
    novovalor = i * 2
    lista.append(novovalor)
print(lista)

#Solução Machado
print(10 * '------')
print('Solução Machado')

#Apliquei aqui uma coisa de operadores que é essa linha 'i *= 2' é a mesma coisa que 'i = i * 2'
#isso funciona para qualquer operação matemática:
# i += 2 --> i = i + 2
# i -= 2 --> i = i - 2
# i /= 2 --> i = i / 2

lista = []
for i in tupla:
  i *= 2
  lista.append(i)
print(lista)
```

```python
# Exercício 4 - Crie uma sequência de números pares entre 100 e 150 e imprima na tela
for i in range(100, 151, 2):
  print(i)


#Solução Machado
print(10 * '------')
print('Solução Machado')

#O que vc fez resolve o problema, pois ele pede os valores entre 100 e 150 que são pares.
#Porém se for solicitado um valor X, tipo 1234567871232, ele é par ou impar?
#Para isso existem os operadores aritméticos (https://www.devmedia.com.br/operadores-no-python/40693), da uma pesquisada nisso tbm.
#Vamos usar o operador '%', onde ele verifica o resto da divisão. TODO número par é divisivel por 2, ou seja, o resto da divisão é 0

#Vou usar um trem chamado f-string, é legal vc estudar ele tbm pq fica mais facil printar os resultados, porém não se preocupe tanto com ele


for i in range(100, 151):
  if i % 2 == 0:
    print(f'{i} é par')
  else:
    print(f'{i} é impar')

#Um desafio legal pra você, peça para o usuário digitar um número e verifique se o número é impar ou par. 
#A função input recebe o valor em string, vc precisa converte-lo para int.
```

```python
## Exercício 5 - Crie uma variável chamada temperatura e atribua o valor 40. Enquanto temperatura for maior que 35, 
# imprima as temperaturas na tela
temperatura= (40)
while temperatura > 35:
  print(temperatura)
  temperatura= temperatura -1

#Solução Machado
print(10 * '------')
print('Solução Machado')

#Aplicando o mesmo conceito de operadores feito no exercicio anterior.
#Vc não precisa colocar o valor entre (), basta digitar 40 que o python vai entender que é um inteiro.

temperatura = 40
while temperatura > 35:
  print(temperatura)
  temperatura -= 1 # temperatura = temperatura - 1
```

```python
# Exercício 6 - Crie uma variável chamada contador = 0. Enquanto counter for menor que 100, imprima os valores na tela,
# mas quando for encontrado o valor 23, interrompa a execução do programa
contador= 0
while contador < 100:
  if contador == 23:
    break
  print(contador)
  contador += 1
  
#Solução correta
```

```python
# Exercício 10 - Faça um programa que conte quantas vezes a letra "r" aparece na frase abaixo. Use um placeholder na 
# sua instrução de impressão

# “É melhor, muito melhor, contentar-se com a realidade; se ela não é tão brilhante como os sonhos, tem pelo menos a 
# vantagem de existir.” (Machado de Assis)

frase = "É melhor, muito melhor, contentar-se com a realidade; se ela não é tão brilhante como os sonhos, tem pelo menos a vantagem de existir."
count=0
for letra in frase:
  if letra == 'r':
    count = count +1
print('A letra r aparece %s vezes na frase!' %(count))

#Solução correta
```
