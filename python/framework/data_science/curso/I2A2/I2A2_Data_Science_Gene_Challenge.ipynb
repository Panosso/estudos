```python
#Read Data
data = pd.read_excel('readcounts.xlsx')
readcounts1 = data
```

```python
roundcount_cp = readcounts1.copy()
roundcount_cp['without_gene'] = roundcount_cp.sum(axis=1)
roundcount_cp = roundcount_cp[roundcount_cp.without_gene != 0]
roundcount_cp.drop('without_gene', axis='columns',inplace=True)
roundcount_cp.set_index(roundcount_cp['Unnamed: 0'])
roundcount_cp.drop('Unnamed: 0', inplace=True, axis=1)
```

```python
names = roundcount_cp.columns
d = preprocessing.normalize(roundcount_cp, axis=0)
roundcount2 = pd.DataFrame(d, columns=names)
```

```python
X = roundcount2.iloc[:,0:].values
kmeans = KMeans(n_clusters=3, init='random')
kmeans.fit(X)
```

```python
atributos = roundcount2.columns
indices = roundcount2.index
cluster_final = pd.DataFrame(X, columns=atributos, index=indices)
cluster_final["family"] = kmeans.labels_

familias = cluster_final["family"].unique()
cluster_final
```

```python
for f in familias:
  individuos = cluster_final[cluster_final["family"] == f]
  print(f"\nFAMILIA {int(f) + 1} : {individuos.index.values}\n")
  print(f"Total de Parentes encontrados : {len(individuos)}\n")
```
